```python
import re
from sudachipy import tokenizer
from sudachipy import dictionary

tokenizer_obj = dictionary.Dictionary().create()

def clean_text(text: str) -> str:
    # text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

def lemmatize_by_sudachipy(sent: str) -> list[str]:
    cleaned_sent = clean_text(sent)
    return [m.dictionary_form() for m in tokenizer_obj.tokenize(cleaned_sent)]

# Example usage
text = "こんにちは、世界！これはテストです。"
lemmatized = lemmatize_by_sudachipy(text)
print(lemmatized)
```

```python
from collections import Counter
with open("lemmas.txt", "r", encoding="utf-8") as file:
    lemmas = file.read().split()

single_letter_hiragana = {
    'あ', 'い', 'う', 'え', 'お',
    'か', 'が', 'き', 'ぎ', 'く', 'ぐ', 'け', 'げ', 'こ', 'ご',
    'さ', 'ざ', 'し', 'じ', 'す', 'ず', 'せ', 'ぜ', 'そ', 'ぞ',
    'た', 'だ', 'ち', 'ぢ', 'つ', 'づ', 'て', 'で', 'と', 'ど',
    'な', 'に', 'ぬ', 'ね', 'の',
    'は', 'ば', 'ぱ', 'ひ', 'び', 'ぴ', 'ふ', 'ぶ', 'ぷ', 'へ', 'べ', 'ぺ', 'ほ', 'ぼ', 'ぽ',
    'ま', 'み', 'む', 'め', 'も',
    'や', 'ゆ', 'よ',
    'ら', 'り', 'る', 'れ', 'ろ',
    'わ', 'を', 'ん'
}

filtered_lemmas = [lemma for lemma in lemmas if lemma not in single_letter_hiragana]
counter = Counter(filtered_lemmas)
```

```python
from collections import Counter
import unicodedata

def is_kanji(char):
    """Check if a character is a kanji."""
    return 'CJK UNIFIED IDEOGRAPH' in unicodedata.name(char)

def extract_kanji_counter(counter):
    """Extract kanji characters from Counter object."""
    kanji_counter = Counter()
    for word, count in counter.items():
        for char in word:
            if is_kanji(char):
                kanji_counter[char] += count
    return kanji_counter

kanji_counter = extract_kanji_counter(counter)
print(kanji_counter)
```

```python
import json

kana_counter = Counter()
kanji_ranks = list(k for k,_ in kanji_counter.most_common())

with open("words.json", "w", encoding="utf-8") as file:
  data = {kanji: [word for word in counter.keys() if kanji in word] for kanji in kanji_ranks}
  for k,v in data.items():
    for w in v:
      if w in kana_counter:
        continue
      kana_counter[w] = 0
  for w in counter.keys():
    if w in kana_counter:
      continue
    kana_counter[w] += 1
  data[None] = [w for w in kana_counter.keys() if kana_counter[w] == 1]

  # usefulness metric
  # for a word => avg rank of kanji components
  # for a kanji => avg metric of words

  metrics = {}
  k_metrics = {}

  for k,v in data.items():
    if len(v) == 0:
      continue
    if k is None:
      continue
    k_metric = 0
    for w in v:
      metric = 0
      _len = 0
      for c in w:
        if c in kanji_ranks:
          metric += kanji_ranks.index(c)
          _len += 1
      metric /= _len
      metrics[w] = metric
      k_metric += metric
    k_metric /= len(v)
    k_metrics[k] = k_metric

  final_data = {}

  sorted_k_metrics = sorted(k_metrics.items(), key=lambda x: x[1])
  # add k from kanji_ranks if not in sorted_k_metrics
  for k in kanji_ranks:
    if k not in sorted_k_metrics:
      sorted_k_metrics.append((k, 0))
  for k,v in sorted_k_metrics:
    final_data[k] = sorted(data[k], key=lambda x: metrics[x])
  final_data[None] = sorted(data[None], key=lambda x: counter[x], reverse=True)

  json.dump(final_data, file, ensure_ascii=False)
```
